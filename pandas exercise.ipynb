### 12. What are the mean scores for students who got A, B, C, F?

```python
mean_list = []
grade_a_students = df[ df["Grade"] == "A"]
total_mean_a = grade_a_students["Total (100)"].mean()
mid_mean_a = grade_a_students["Mid"].mean()
mid_mean_a_percent = mid_mean_a / 30 * 100

print(f"total mean: {total_mean_a} -- mid mean: {mid_mean_a_percent}")
```

```python
mean_list = []
grade_b_students = df[ df["Grade"] == "B"]
total_mean_b = grade_b_students["Total (100)"].mean()
mid_mean_b = grade_b_students["Mid"].mean()
mid_mean_b_percent = mid_mean_b / 30 * 100

print(f"total mean: {total_mean_b} -- mid mean: {mid_mean_b_percent}")
```

```python
mean_list = []
grade_c_students = df[ df["Grade"] == "C"]
total_mean_c = grade_c_students["Total (100)"].mean()
mid_mean_c = grade_c_students["Mid"].mean()
mid_mean_c_percent = mid_mean_c / 30 * 100

print(f"total mean: {total_mean_c} -- mid mean: {mid_mean_c_percent}")
```

```python
df_mean = pd.DataFrame(columns=["Grade","Total", "Mids"])
df_mean
```

```python
df_mean = df_mean.append({"Grade":"A", "Total":total_mean_a, "Mids": mid_mean_a_percent}, ignore_index=True)
df_mean = df_mean.append({"Grade":"B", "Total":total_mean_b, "Mids": mid_mean_b_percent}, ignore_index=True)
df_mean = df_mean.append({"Grade":"C", "Total":total_mean_c, "Mids": mid_mean_c_percent}, ignore_index=True)
```
